```python
 #import the required liabraries 
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd    
```

```python
#To view the plot in notebook
%matplotlib inline
```

```python
#import the auto data set
df = pd.read_csv("auto_data_mpg.csv")
```

```python
#view the top 5 records
df.head()
```

```python
#write a user defined function for origin
# 1-USA, 2-Europe, 3-Asia
def origin(num):
    if num ==1:
        return "USA"
    elif num==2:
        return "Europe"
    elif num==3:
        return "Asia"
#use apply function
df['origin'] = df['origin'].apply(origin)
```

```python
#view first 30 data points after applying the function
df.head(30)
```

```python
#draw the pair plot using sns for mpg, weight, origin and with hue origin,set the size to 4
# note: hue is variable in data set to map plot aspects to different colours
sns.pairplot(df[['mpg','weight','origin']])
```

```python
sns.pairplot(df[['mpg','weight','origin']],hue ='origin',size=4)
```
