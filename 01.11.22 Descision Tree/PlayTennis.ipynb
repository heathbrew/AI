```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score
```

```python
dataset = pd.read_csv(r"PlayTennis.csv")
le = LabelEncoder()
dataset["outlook"] = le.fit_transform(dataset["outlook"])
dataset["temp"] = le.fit_transform(dataset["temp"])
dataset["humidity"] = le.fit_transform(dataset["humidity"])
dataset["windy"] = le.fit_transform(dataset["windy"])

y = le.fit_transform(dataset["play"])
x = dataset.drop(["play"],axis=1)

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.4)
```

```python
model = DecisionTreeClassifier(criterion="gini")
model.fit(x,y)
```

```python
y_pred = model.predict(x_test)
acs = accuracy_score(y_test,y_pred)
acs*100
```

```python
from sklearn import tree
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(20,100))
tree.plot_tree(model,fontsize=10)
plt.show()
```
