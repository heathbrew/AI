Problem Statement

You have been provided with a dataset that lists Ohio State's leading causes of death from the year 2012 .

Using the two data points cause of death and percentile , draw a pie chart to visualize the
dataset .

```python
#import required liabraries
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Data from Ohio state , Leading causes of desth in the state form year 2012
#cause of death
cause = 'Chronic Diseases','Unintentiional Injuries','Alzheimers' , 'Influenza and Pneumonia','Sepsis','Others'
#percentile
percentile = [62,5,4,2,1,26]
```

```python
#set the pie chart plot properties
#set figure size
plt.figure(figsize=(10,10))
#explode the largest pie in the data set
explode = (0.5,0,0,0,0,0)
#set pie chart properties
plt.pie(percentile,labels=cause,explode=explode,autopct='%1.1f%%',startangle=70)
#set axis equal to draw pie as circle 
plt.axis('equal')
#set title of the pie chart
plt.title('Ohio State -2012 : Leading Causes of Death')
#Show the plot (pie chart)
plt.show()
```
